Sliding window诀窍：
- 通常先移动右指针，再移动左指针缩小。通常需要以空间换时间，所以需要用到dict记录当前状态。
- 通常用来找极值。如果遇到要找所有情况的题目，需要把题目进行转换
- 在找符合条件的permutation的时候，通常先移动左指针再从左指针的问题移动右指针来确定情况
- 通常需要一个dict来记录当前状态。 少一个就减1，多一个就加1
- 如果在加1 之后，dict当前记录==1，则说明增加了一个新的元素。反之则没有
- 如果在减1之后dict当前记录<=0则说明移除了一个元素，反之则没有
- 移动右指针直到不符合条件
- 再移动左指针直到符合条件
- 如果是找permutation的题目通常左右各移动一次来滑动寻找

# Minimum Window Substring

```python
class Solution:
    def minWindow(self, s: str, t: str) -> str:

        # Create a hashmap to record frequency in t
        hashmap = {}
        for i in t:
            if i in hashmap:
                hashmap[i] += 1
            else:
                hashmap[i] = 1
        # Create a counter to record number of matches in t
        counter = 0
        n_t = len(t)
        n_s = len(s)
        min_len = n_s + 1
        l = 0
        min_string = ""
        # Loop through s use the right pointer
        for r in range(n_s):
            # print(f'{r=}')
             # for each step
            # if s[r] in hashmap then hashmap[s[r]] - 1
            if s[r] in hashmap:
                 hashmap[s[r]] -= 1
                 # if hashmap[s[r]] still >= 0 means it is a useful letter, so counter += 1
                 if hashmap[s[r]] >= 0:
                    counter += 1
            # stop moving r when counter == len(t)
            # now start to move l
            while counter == n_t:
                if r - l + 1 < min_len:
                    # update the length and min string
                    min_len = r - l + 1
                    min_string = s[l:r+1]
                # if hashmap[s[l]] + 1 still >= 0 means the substring can still shrink by moving l
                if s[l] not in hashmap:
                    l = l + 1
                else:
                    hashmap[s[l]] = hashmap[s[l]] + 1
                    if hashmap[s[l]]> 0:
                        counter -= 1
                    l = l + 1
                # print(f'{l=}')
                # print(f'{counter=}')


        return min_string
```

# Longest Substring Without Repeating Characters

```python
def lengthOfLongestSubstring(self, s: str) -> int:
        # two pointer l, r
        # for l in range(len(s)):
        # for each s[l] find the longest subtring
        # map to record the current state
        # when i move r, if the count > 1, I need to move left until count = 1.
        # move r until count >= 1
        # move l until count = 1
        # build the map
        # corner case
        if len(s) <= 1:
            return len(s)

        hashmap = {}
        n_s = len(s)
        max_len = 0
        l = 0
        for r in range(n_s):
            if hashmap.get(s[r], 0) == 1:
                max_len = max(max_len, r - l)
            if s[r] in hashmap:
                hashmap[s[r]] += 1
            else:
                hashmap[s[r]] = 1
            
            while hashmap[s[r]] > 1:
                # move l
                hashmap[s[l]] -= 1
                l = l + 1
        # boundary
        max_len = max(max_len, r - l + 1)

        return max_len
```

# Longest Repeating Character Replacement

```python
def characterReplacement(self, s: str, k: int) -> int:
        # counter store number of operation
        # max_length to store the max length
        # left, right pointer
        # move the right pointer first
        # at every step record the current letter counts
        # if the current letter counts larger than max, then update max cnts
        # max count + k should be the current window length
        # if the max count + k > than the max length then record the max length
        # move left pointer 
        # if the max cnt letter changes then start the move the right pointer again

        # repeat the above steps until going through the whole string


        max_length = 1
        letter_cnt = {}
        cur_max_cnt = 0
        l = 0
        for r in range(len(s)):
            if s[r] not in letter_cnt:
                letter_cnt[s[r]] = 1
            else:
                letter_cnt[s[r]] += 1
            # update the current max letter cnt and max letter
            if letter_cnt[s[r]] >= cur_max_cnt:
                cur_max_cnt = letter_cnt[s[r]]
                max_letter = s[r]


            # check if max count + k <= current window length
            while cur_max_cnt + k < r - l + 1:
                # move the left
                letter_cnt[s[l]] -= 1
                l += 1
                # 为什么这里移动了左指针却不需要更新max count呢
                # 因为这里是求最长的子串，max count相当于卡了一个窗口
                # 只有出现更大的max count的时候才有可能出现更长的子串
                # cur_max_cnt = max(letter_cnt.values())
                
                
            max_length = max(r-l+1, max_length)


        return max_length
```

 # Max Consecutive Ones III


```python
class Solution:
    def longestOnes(self, nums, k) -> int:
        # two pointer sliding window
        # move the right pointer until ones count + K < r - l + 1
        # then move the left pointer unitl ones count + K < r - l + 1

        max_len = 0
        ones_cnt = 0
        l = 0
        for r in range(len(nums)):
            if nums[r] == 1:
                ones_cnt += 1

            while ones_cnt + k < r - l + 1:
                if nums[l] == 1:
                    ones_cnt -= 1
                l += 1

            max_len = max(max_len, r-l+1)

        return max_len
```

# Permutation in String

```python
# 用固定size的sliding window来扫描。
# 每次移动左指针和右指针各一步，然后根据左右指针的移动来更新temp_dict
# 直接比较 temp_dict 和目标dict
class Solution:
    def checkInclusion(self, s1: str, s2: str) -> bool:
        # sliding window scan through s2
        # dict record s1
        # each step move left point once and right point once
        # each step update the temp dict
        # compare temp dict with s1 dict
        def counter(s):
            s_dict = {}
            for i in s:
                s_dict[i] = s_dict.get(i, 0) + 1
            return s_dict

        l = 0
        n_s1 = len(s1)
        r = n_s1 - 1
        n_s2 = len(s2)
        s1_dict = {}

        s1_dict = counter(s1)
        temp_dict = counter(s2[l:r+1])
        if s1_dict == temp_dict:
            return True
        for l in range(0,n_s2-n_s1):
            # print(f'{temp_dict}')
            temp_dict[s2[l]] = temp_dict.get(s2[l], 0) - 1
            if temp_dict[s2[l]]<=0:
                temp_dict.pop(s2[l])
            r = r + 1
            temp_dict[s2[r]] = temp_dict.get(s2[r], 0) + 1
            if s1_dict == temp_dict:
                return True
        else:
            return False
```

# Find All Anagrams in a String

```python
def findAnagrams(self, s: str, p: str):
        def counter(s):
            dict_s = {}
            for i in s:
                dict_s[i] = dict_s.get(i,0) + 1
            return dict_s

        l = 0
        n_p = len(p)
        n_s = len(s)
        dict_p = counter(p)
        dict_temp = counter(s[0:n_p])
        res = []
        if dict_temp == dict_p:
            res.append(l)
        # boundary condition is the difficult part
        # walk through the process in mind
        for l in range(0, n_s-n_p):
            # move l first
            l_cur = s[l]
            dict_temp[l_cur] = dict_temp.get(l_cur,0) -1
            if dict_temp[l_cur]<=0:
                dict_temp.pop(l_cur)

            # then move r
            r = l + n_p
            r_cur = s[r]
            dict_temp[r_cur] = dict_temp.get(r_cur,0) + 1

            if dict_temp == dict_p:
                res.append(l+1)

        return res
```

# Subarrays with K Different Integers


sliding window方法通常是用来寻找极值的。比方说最长的，最短的，最多包涵，最少包涵。在找极值的时候，双指针单向移动不检测太小或者太大的目标都不会影响搜索结果。但是这道题需要找出所有有k个整数的情况。由于双指针无法回头，这里需要把问题变化成找极值的问题。事实上刚好有k个整数的情况 = 最多有k个整数的情况 - 最多有k-1个整数的情况。

```python



def subarraysWithKDistinct(self, nums: List[int], k: int) -> int:

        # two pointer can't solve find all possiblity problem
        # because the pointers only moves in one direction, and never turn back
        # this problem we have to convert it into two sub problems
        # one is find the arrays with most k element
        # two is find the arrays with most k- 1 elements
        # one minus two would give us the anwoer
        # find the k most is solvable by the sliding window

        # move the right pointer until find k elements
        # move the left pointer until until elemnts = k - 1

        def get_most_k(nums, k):
            mapping = {}
            distinct_cnt = 0
            res = 0
            l = 0
            for r in range(len(nums)):
                mapping[nums[r]] = mapping.get(nums[r],0)+1
                if mapping[nums[r]] == 1:
                    distinct_cnt += 1
                # 这里的关键是先移动右指针，每次移动都把符合条件<=k的subarray算上
                if distinct_cnt <= k:
                    res += r - l + 1
                else:
                    # 当总的integer > k时，因为不满足条件而开始移动左指针
                    while distinct_cnt > k:
                        mapping[nums[l]] = mapping.get(nums[l],0) -1
                        if mapping[nums[l]] <= 0:
                            distinct_cnt -= 1
                        l += 1
                    # 当左指针移动到将将满足条件的时候。这个时候因为右指针有可能和左指针不是重合的。所以要算上当前左右之前可能的subarray
                    res += r - l + 1


            return res

        # print(get_most_k(nums,k))
        # print(get_most_k(nums, k-1))

        return get_most_k(nums,k) - get_most_k(nums, k-1)
```
