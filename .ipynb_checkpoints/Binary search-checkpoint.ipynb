# Binary search

Binary search是一种针对有序区间O(logN)搜索方式，最常见用于已经排好序的Array


## Binary search 两大基本原则
1. Shrink the search space every iteration
2. Cannot exclude potential answers during each shrinking

## 三大模板
1. 找一个准确值
循环条件： l<=r (左边都移动才用小于等于）
缩减空间： l = mid + 1, r = mid - 1

2. 找一个模糊值（比4大的最小数）
循环条件：l<r （搜索条件要注意最后还剩下两个数的时候，是不是会无线循环下去） （只移动一边才用小于）
缩减空间：l = mid, r = mid - 1或者l = mid + 1, r = mid
（每个iteration最好把mid > value, mid = value 和 mid < value 三种情况都列出来仔细思考）

3. 万用型
循环条件： l< r - 1
缩减搜索空间： l = mid, r = mid

# 34. Find First and Last Position of Element in Sorted Array

Given an array of integers nums sorted in non-decreasing order, find the starting and ending position of a given target value.

If target is not found in the array, return [-1, -1].

You must write an algorithm with O(log n) runtime complexity.

```python
# first find the starting point, then use the starting point to find the ending point

l = 0
r = len(nums) - 1

if target not in nums:
    return [-1, -1]



while l < r:
    mid = l + (r - l)//2
    if nums[mid] == target:
        r = mid

    elif nums[mid] > target:
        r = mid - 1

    else:
        l = mid + 1


start = r
end = start

while end < len(nums) and nums[end] == target:
    end += 1

return [start, end - 1]
```

# 81. Search in Rotated Sorted Array II

There is an integer array nums sorted in non-decreasing order (not necessarily with distinct values).

Before being passed to your function, nums is rotated at an unknown pivot index k (0 <= k < nums.length) such that the resulting array is [nums[k], nums[k+1], ..., nums[n-1], nums[0], nums[1], ..., nums[k-1]] (0-indexed). For example, [0,1,2,4,4,4,5,6,6,7] might be rotated at pivot index 5 and become [4,5,6,6,7,0,1,2,4,4].

Given the array nums after the rotation and an integer target, return true if target is in nums, or false if it is not in nums.

You must decrease the overall operation steps as much as possible.

```python
# 方法1：找到旋转点， 对两边分别进行二分查找 
def search(self, nums, target):
    """
    :type nums: List[int]
    :type target: int
    :rtype: bool
    """


    if len(nums) < 2:
        return target == nums[0]

    def search_num(nums, target):
        l = 0
        r = len(nums) - 1

        while l <= r:
            mid = l + (r - l)//2
            if nums[mid] == target:
                return True
            elif nums[mid] > target:
                r = mid - 1
            else:
                l = mid + 1

        return False


    pivot = 1
    for i in range(len(nums) - 1):
        if nums[i] > nums[i + 1]:

            pivot = i + 1


    output = search_num(nums[:pivot], target) or search_num(nums[pivot:], target)

    return output 
```

```python
# 方法2：即使数组被旋转过，我们仍然可以利用这个数组的递增性，使用二分查找。对于当前的中点，
# 如果它指向的值小于等于右端，那么说明右区间是排好序的；反之，那么说明左区间是排好序的。
# 如果目标值位于排好序的区间内，我们可以对这个区间继续二分查找；反之，我们对于另一半区
# 间继续二分查找。
# 注意，因为数组存在重复数字，如果中点和左端的数字相同，我们并不能确定是左区间全部
# 相同，还是右区间完全相同。在这种情况下，我们可以简单地将左端点右移一位，然后继续进行
# 二分查找。

def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: bool
        """
        
        pivot = 1
        
        for i in range(len(nums) - 1):
            if nums[i] > nums[i+1]:
                pivot = i
                
                
        l = 0
        r = len(nums) - 1
        
        while l<=r:
            
            mid = l + (r - l)//2
            
            if nums[mid] == target:
                return True
            elif mid <= pivot:
                if nums[mid] > target and nums[l] <= target:
                    r = mid -  1
                else:
                    l = mid + 1
                    
            else:
                if nums[mid] < target and nums[r] >= target:
                    l = mid + 1
                else:
                    r = mid - 1
                    
        
        return False
        
```

# 540. Single Element in a Sorted Array

You are given a sorted array consisting of only integers where every element appears exactly twice, except for one element which appears exactly once.

Return the single element that appears only once.

Your solution must run in O(log n) time and O(1) space.

```python
def singleNonDuplicate(self, nums):
    """
    :type nums: List[int]
    :rtype: int
    """
    # 判断一个数是否左右有重复， 没有的话，输出结果
    # 如果有重复的数，找到重复的两个数，看左边和右边长度是奇还是偶数。答案肯定是长度为奇数的那一边

    l = 0
    r = len(nums) - 1


    while l <= r:

        mid = l + (r - l)//2

        if mid - 1 > -1 and nums[mid-1] == nums[mid]:
            mid_1 = mid - 1
            mid_2 = mid


            if (mid_1 - l) % 2 == 0:
                l = mid_2 +  1
            else:
                r = mid_1 - 1

        elif mid + 1 < len(nums) and nums[mid+1] == nums[mid]:
            mid_1 = mid
            mid_2 = mid + 1

            if (mid_1 - l) % 2 == 0:
                l = mid_2 + 1
            else:
                r = mid_1 - 1

        else:
            return nums[mid]
```

# 35. Search Insert Position

Given a sorted array and a target value, return the index if the target is found. If not, return the index where it would be if it were inserted in order.

You may assume no duplicates in the array.

Tips: 这道题用二分法搜索解答

时间复杂度 O(logn)
空间复杂度 O(1)

```python
class Solution(object):
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        N = len(nums)
        # 值得注意的是right的初始值是N，而不是N-1.因为如果right设置成N-1.而nums的长度只有1的话。一开始就无法执行while loop
        left, right = 0, N
        
        while left < right:
            # 关键部分在于求mid的时候一定要用下取整（//），否则会出现小数。因为这里使用了下取整，所以当left和right相邻的时候，
            # left就会自动变成 mid。解决方案如下
            mid = left + (right-left)//2
            if nums[mid] == target:
                return mid
            elif nums[mid]> target:
                right = mid
            # 当 mid要小于target的时候，left直接设置成mid+1。这里可能担心如果mid+1 大于target怎么办。 其实
            # 如果mid+1是刚好大于target的，那最后二分搜索到只剩两个元素的时候，也会找到left,因为下取整的关系。
            else:
                left = mid+1
        return left
```

```python
# 这里是另外一种标准写法，是用二分法搜索一个数
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        
        N = len(nums)
        left, right = 0, N-1
        
        while left<=right:
            mid = left+(right-left)//2
            
            if nums[mid]==target:
                return mid
            elif nums[mid]>target:
                right = mid-1
            elif nums[mid]< target:
                left = mid + 1
                    
        return left
```

# 34. Find First and Last Position of Element in Sorted Array

Given an array of integers nums sorted in ascending order, find the starting and ending position of a given target value.

Your algorithm's runtime complexity must be in the order of O(log n).

If the target is not found in the array, return [-1, -1].

Tips: 这道题和上面得题目类似。不同点在于，target一定要存在array中才有解，否则就是[-1,-1].

```python
class Solution(object):
    def searchRange(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        N = len(nums)
        left, right = 0, N
     
        while left < right:
            mid = left + (right-left)//2
            # 当在array中找到解的时候，必须有一个输出。所以在这个if下面有return. 并执行以下操作
            if nums[mid] == target:
                # 先把left和right都换到mid,然后分别往两边一格一格移动，只要还能找到和target一样的值就继续移动
                left, right = mid, mid
                while right<N-1 and nums[right+1] == target:
                    right +=1
                while left > 0 and nums[left-1] == target:
                    left -=1
                return [left,right]
            elif nums[mid]>target:
                right = mid
            else:
                left = mid+1
                
        return [-1, -1]
```

```python
class Solution(object):
    def searchRange(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        # 标准的左侧边界解法
        N = len(nums)
        left,right = 0, N-1
        # Find the left boundary
        while left<=right:
            mid = left +(right-left)//2
            if nums[mid]==target:
                right = mid -1
            elif nums[mid]> target:
                right = mid -1
            elif nums[mid] < target:
                left = mid +1
        # 因为是找边界，最后判定边界left是否出界以及最后left停在的位置target是否找到        
        if left>= N or nums[left]!=target:
            return [-1,-1]
        left_bound = left
        
        # 从这里开始时标准的右侧边界解法
        left,right = 0, N-1
        # Find the right boundary
        while left <= right:
            mid = left +(right-left)//2
            if nums[mid]==target:
                left = mid +1
            elif nums[mid]>target:
                right = mid -1
            elif nums[mid]< target:
                left = mid +1
                
        if right<0 or nums[right]!=target:
            return [-1, -1]
        
        right_bound = right
        
        return[left_bound, right_bound]
        
```

# 704. Binary Search

Given a sorted (in ascending order) integer array nums of n elements and a target value, write a function to search target in nums. If target exists, then return its index, otherwise return -1.

Tips: 这题是easy,和上两题的解法基本类似

```python
class Solution(object):
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        nums.sort()
        
        N = len(nums)
        left, right = 0, N
        
        while left<right:
            mid = left + (right-left)//2
            if nums[mid] == target:
                return mid
            elif nums[mid]>target:
                right = mid
            else:
                left = mid +1
        
        return -1
```

# 981. Time Based Key-Value Store

Create a timebased key-value store class TimeMap, that supports two operations.

1. set(string key, string value, int timestamp)

Stores the key and value, along with the given timestamp.
2. get(string key, int timestamp)

Returns a value such that set(key, value, timestamp_prev) was called previously, with timestamp_prev <= timestamp.
If there are multiple such values, it returns the one with the largest timestamp_prev.
If there are no values, it returns the empty string ("").

Tips: 这道题涉及数据结构和二分法的搜索

```python
import collections as ct
class TimeMap(object):

    def __init__(self):
        """
        Initialize your data structure here.
        """
     
        # defaultdict()是非常灵活实用的数据结构。Key可以是任何东西，value必须list.如果不提供list,会生成一个默认值（？）
        
        self.i_map = ct.defaultdict(list)

    def set(self, key, value, timestamp):
        """
        :type key: str
        :type value: str
        :type timestamp: int
        :rtype: None
        """
    
        # {v:value,ts:timestamp}是list里面的一个元素，每一个key对应一个list.这样做的好处是在input一个key的时候，会自动把其他key的
        # list（信息）给筛除
        self.i_map[key].append({'v':value, 'ts':timestamp})


    def get(self, key, timestamp):
        """
        :type key: str
        :type timestamp: int
        :rtype: str
        """
        # 传统的找右侧边界
        arr = self.i_map[key]
        N = len(arr)
        left,right = 0,N-1
        while left <= right:
            mid = left + (right-left)//2
            #.ts .后面必须是string
            if arr[mid]['ts'] == timestamp:
                left = mid + 1
            elif arr[mid]['ts']> timestamp:
                right = mid -1
            elif arr[mid]['ts'] < timestamp:
                left = mid + 1       
        # 当函数中不存在target的时候，右侧边界输出的是刚好小于或者等于target的最右。(while loop结束的时候，right=left-1)反推可得：左侧边界则是大于或者等于target得最左。
        # if判定当right出界的时候，output ""。这里最后输出的结构不需要等于target，所以arr[right]['ts']!=target的这个条件是不需要的
        if right < 0:
            return ""
        else:
            return arr[right]['v']
```
